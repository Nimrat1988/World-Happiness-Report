2. Use the questions you defined in the previous task to pick out variables from your data set suitable for your exploratory visual analysis.

```python
#Selecting only numerical columns
df_numeric = df[['year', 'life_ladder', 'gdp_per_capita', 'social_support', 'life_expectancy_birth', 'freedom_for_life_choices', 'generosity', 'perceptions_of_corruption', 'positive_affect', 'negative_affect']]
```

```python
#Creating a correlation matrix
corr_matrix = df_numeric.corr()
```

```python
#Plotting correlation matrix
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Happiness Indicators", fontsize=16)
plt.show()
```

```python
#Export heatmap
plt.savefig(os.path.join(path, 'Analysis', 'heatmap_world_happiness_report.png'))
```

```python
#Creating scatterplot for life expectancy vs GDP per capita
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="gdp_per_capita", y="life_expectancy_birth", alpha=0.6)

plt.title("Life Expectancy vs GDP per Capita")
plt.xlabel("GDP per Capita (Log Scale)")
plt.ylabel("Healthy Life Expectancy at Birth")
plt.grid(True, linestyle="--", alpha=0.3)
plt.show()
```

Scatterplot shows a strong positive relationship between GDP per capital and healthy life expectancy. As a country's income level increases, its population tends to live longer.

```python
#Export scatterplot
plt.savefig(os.path.join(path, 'Analysis', 'scat1_world_happiness_report.png'))
```

```python
#Creating scatterplot for GDP per capita vs Life Ladder
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='gdp_per_capita', y='life_ladder')
plt.title('GDP per Capita vs Life Ladder')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Ladder (Happiness Score)')
plt.show()
```

Scatterplot shows a clear upward trend. Countries with higher income tend to report higher happiness scores.

```python
#Export scatterplot
plt.savefig(os.path.join(path, 'Analysis', 'scat2_world_happiness_report.png'))
```

```python
#Creating scatterplot for Social support vs Life Ladder
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='social_support', y='life_ladder')
plt.title('Social Support vs Life Ladder')
plt.xlabel('Social Support')
plt.ylabel('Life Ladder')
plt.show()
```

This relationship is also strongly positive. Countries where people feel they have someone to rely on tend to be happier.

```python
#Export scatterplot
plt.savefig(os.path.join(path, 'Analysis', 'scat3_world_happiness_report.png'))
```

```python
#Creating a scatterplot for Perception of corruption vs Life ladder
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='perceptions_of_corruption', y='life_ladder')
plt.title('Perceptions of Corruption vs Life Ladder')
plt.xlabel('Perceptions of Corruption')
plt.ylabel('Life Ladder')
plt.show()
```

The scatterplot shows a downward trend. Higher corruption perceptions are linked with lower happiness levels.

5. Create a pair plot of the entire data set.
Comment on the distribution of the variables and mark variables youd like to explore further with an explanation of why.

```python
#Creating a pair plot of the entire dataset
sns.pairplot(df_numeric, diag_kind="kde", corner=True)
plt.suptitle("Pair Plot of Happiness Dataset", y=1.02, fontsize=14)
plt.show()
```

The pair plot helps visualize the distribution of numerical variables and their relationships. Variables such as Log GDP per capita, Social support, Life expectancy, and Freedom to make life choices show strong positive relationships with Life Ladder (happiness score). These plots show upward-sloping patterns, indicating that higher economic growth, better social networks, longer life expectancy, and greater personal freedoms all contribute to higher happiness levels. Negative affect shows a negative trend with happiness, suggesting that emotional distress reduces well-being. These strongly correlated variables will be explored further, as they appear most relevant for understanding what drives happiness across countries.

```python
#Export pair plot
plt.savefig(os.path.join(path, 'Analysis', 'pairplot_world_happiness_report.png'))
```

6. Create a categorical plot and interpret the results.

```python
#Creating categorial plot for Average Happiness by Year
plt.figure(figsize=(12,6))
sns.barplot(data=df, x="year", y="life_ladder", palette="coolwarm")
plt.title("Average Global Happiness by Year")
plt.xlabel("Year")
plt.ylabel("Average Life Ladder")
plt.xticks(rotation=45)
plt.show()
```

The highest average score appears in 2005, followed by a noticeable dip in 2006. From 2007, the global average happiness score stayed relatively consistent with only minor fluctuations. Overall, chart shows long-term stability in global happiness, with only modest increases or decreases over time. 

```python
#Exporting categorial plot
plt.savefig(os.path.join(path, 'Analysis', 'cat_plot_world_happiness_report.png'))
```
