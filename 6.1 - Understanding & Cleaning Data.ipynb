```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
#Project folder path as a string
path = r'/Users/Nimrat/Documents/World Happiness Report'
```

```python
#Import the “world happiness report.csv” data set into your notebook using the os library
df = pd.read_csv(os.path.join(path, 'Original data', 'world_happiness_report.csv'), index_col = False, encoding='latin1')
```

```python
#To print data structure
print(df.info())
```

```python
#To check the values of the data
df.head(10)
```

```python
# To check missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# To standardize column names by removing extra spaces
df = df.rename(columns={
    'Country name': 'country_name',
    'Life Ladder': 'life_ladder',
    'Log GDP per capita': 'gdp_per_capita',
    'Social support': 'social_support',
    'Healthy life expectancy at birth': 'life_expectancy_birth',
    'Freedom to make life choices': 'freedom_for_life_choices',
    'Generosity': 'generosity',
    'Perceptions of corruption': 'perceptions_of_corruption',
    'Positive affect': 'positive_affect',
    'Negative affect': 'negative_affect'})
```

```python
#To check the output
df.head()
```

```python
#To fill missing values with mean
cols_to_fill = ['gdp_per_capita', 'social_support', 'life_expectancy_birth', 'freedom_for_life_choices', 'perceptions_of_corruption', 'positive_affect', 'negative_affect']
df[cols_to_fill] = df[cols_to_fill].fillna(df[cols_to_fill].mean())
```

```python
#To fill missing values in generosity column with median as there are positive and negative values.
df['generosity'] = df['generosity'].fillna(df['generosity'].median())
```

```python
#To check the output
df.isnull().sum()
```

```python
df.head()
```

```python
#To check for duplicates
df.duplicated().sum()
```

```python
#To perform basic descriptive statistical analysis
df.describe()
```

```python
#Export the “world happiness report.csv” cleaned data set into your notebook using the os library
df.to_csv(os.path.join(path, 'Prepared data', 'world_happiness_report_cleaned.csv'))
```
